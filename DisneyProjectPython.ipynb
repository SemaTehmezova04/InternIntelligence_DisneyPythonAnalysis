4. Perform Calculations

```python
# Total number of titles
import pandas as pd

# Make sure to replace 'disney_plus_titles.csv' with the correct file path
file_path = 'disney_plus_titles.csv'

# Load the dataset
df = pd.read_csv(file_path)

# Check if the dataframe is loaded properly
if df is not None:
    total_titles = df.shape[0]
    print(f"Total number of titles: {total_titles}")
else:
    print("The dataset could not be loaded.")

```

```python
# Count by type (Movies vs. TV Shows)
type_counts = df['type'].value_counts()
print("\nTitles by Type:\n", type_counts)
```

```python
# Most popular genres
top_genres = df['listed_in'].value_counts().head(10)
print(top_genres)
```

```python
# Count by rating
rating_counts = df['rating'].value_counts()
print("\nCount by Rating:\n", rating_counts)
```

```python
# Add a 'decade' column based on release year
df['release_decade'] = (df['release_year'] // 10) * 10
decade_counts = df['release_decade'].value_counts().sort_index()
print("\nTitles by Decade:\n", decade_counts)
```

5. Visualize Key Insights

```python
# Bar chart: Titles by Type
sns.barplot(x=type_counts.index, y=type_counts.values, palette='crest')
plt.title('Titles by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart: Titles by Type
sns.barplot(x=type_counts.index, y=type_counts.values, palette='crest')
plt.title('Titles by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart: Distribution of Ratings
rating_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title='Distribution of Ratings')
plt.ylabel('')  # This removes the y-axis label
plt.show()
```

```python
# Line chart: Titles Released Over Decades
decade_counts.plot(kind='line', marker='o', title='Titles Released Over Decades')
plt.xlabel('Decade')
plt.ylabel('Number of Titles')
plt.grid()
plt.show()
```
